```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATA
runs = [10, 20, 30]
datasets = ['X-Sub', 'X-View']

x_rollout_10 = [10, 20, 30]
x_rollout_20 = [20, 40]
x_rollout_30 = [30, 60]

xsub_accuracy = {
    '10': [80.1, 85.9, 85.9],
    '20': [88.3, 81.5],
    '30': [86.4, 82.1]
}

xview_accuracy = {
    '10': [95.3, 94.5, 91.4],
    '20': [95.1, 94.7],
    '30': [92.8, 95.5]
}

xview_average = {
    '10': [71.3, 72, 68.6],
    '20': [56.2, 66.4],
    '30': [69.1, 60.0]
}

xsub_average = {
    '10': [50.4, 61.3, 62.3],
    '20': [45.7, 33.5],
    '30': [49.6, 60.3]
}
```

```python
# MAX values
xsub_max_accuracy = max((max(values), int(key)) for key, values in xsub_accuracy.items())
print("Maximum accuracy:", xsub_max_accuracy[0])
print("Corresponding key:", xsub_max_accuracy[1])

xview_max_accuracy = max((max(values), int(key)) for key, values in xview_accuracy.items())
print("Maximum accuracy xview:", xview_max_accuracy[0])
print("Corresponding key:", xview_max_accuracy[1])
```

```python
# PLOT
# Define colors for each group of runs
font = {'fontname': 'Times New Roman', 'size'   : 12}
colors = {'10': 'blue', '20': 'green', '30': 'purple'}
# Plotting xsub dataset
plt.figure(figsize=(6, 3.7))
# plt.title('X-Sub dataset rollout influence', **font)
plt.xlabel('# Iterations', **font)
plt.ylabel('Top-1 Accuracy (%)', **font)
for run, x_values in zip(['10', '20', '30'], [x_rollout_10, x_rollout_20, x_rollout_30]):
    plt.plot(x_values, xsub_accuracy[run], 'o', label=f'{run} rollouts before update', color=colors[run])
    # max_index = xsub_accuracy[run].index(max_accuracy)
    # plt.scatter(x_values[max_index], max_accuracy, s=100, color='red', marker='o', zorder=5)
plt.scatter(xsub_max_accuracy[1], xsub_max_accuracy[0], s=200, color='green', marker='*', zorder=10)
plt.xticks([10, 20, 30, 40, 60])
plt.legend(prop={"family":"Times New Roman"})
plt.grid(True)
plt.savefig('./tempe4.pdf', bbox_inches = 'tight')

# Plotting xview dataset
plt.figure(figsize=(6, 3.7))
#plt.title('X-View dataset rollout influence', **font)
plt.xlabel('# Iterations', **font)
plt.ylabel('Top-1 Accuracy (%)', **font)
for run, x_values in zip(['10', '20', '30'], [x_rollout_10, x_rollout_20, x_rollout_30]):
    plt.plot(x_values, xview_accuracy[run], 'o', label=f'{run} rollouts before update', color=colors[run])


plt.scatter(60, xview_max_accuracy[0], s=200, color='purple', marker='*', zorder=10)
plt.xticks([10, 20, 30, 40, 60])
plt.legend(prop={"family":"Times New Roman"})
plt.grid(True)
plt.savefig('./tempe3.pdf', bbox_inches = 'tight')
```

```python
import matplotlib.pyplot as plt

# Define colors for each group of runs
font = {'fontname': 'Times New Roman', 'size': 12}
colors = {'10': 'blue', '20': 'green', '30': 'purple'}
marker = {'10': '>', '20': 'o', '30': 'P'}

fig, axes = plt.subplots(1, 2, figsize=(12, 3.7))

# Plotting xsub dataset
plt.subplot2grid((1, 2), (0, 0))
plt.xlabel('# Iterations', **font)
plt.ylabel('Top-1 Accuracy (%)', **font)
for run, x_values in zip(['10', '20', '30'], [x_rollout_10, x_rollout_20, x_rollout_30]):
    plt.plot(x_values, xsub_accuracy[run], marker[run], markersize=8, label=f'{run} rollouts preceding update', color=colors[run])
plt.scatter(xsub_max_accuracy[1], xsub_max_accuracy[0], s=250, color='green', marker='*', zorder=12)
plt.xticks([10, 20, 30, 40, 60])
plt.title('X-Sub rollout influence', **font)
plt.legend(prop={"family": "Times New Roman"})
plt.grid(True)

# Plotting xview dataset
plt.subplot2grid((1, 2), (0, 1))
plt.xlabel('# Iterations', **font)
plt.ylabel('Top-1 Accuracy (%)', **font)
for run, x_values in zip(['10', '20', '30'], [x_rollout_10, x_rollout_20, x_rollout_30]):
    plt.plot(x_values, xview_accuracy[run], marker[run], markersize=9, label=f'{run} rollouts preceding update', color=colors[run])
plt.scatter(60, xview_max_accuracy[0], s=250, color='purple', marker='*', zorder=12)
plt.xticks([10, 20, 30, 40, 60])
plt.title('X-View rollout influence', **font)
plt.legend(prop={"family": "Times New Roman"})
plt.grid(True)

plt.tight_layout()
plt.savefig('./tempe3.pdf', bbox_inches='tight')
plt.show()
```
